```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
iris=load_iris()
```

```python
iris
```

```python
iris.keys()
```

```python
x=iris.data[:,:-2]
y=iris.target
```

```python
iris.target_names
```

```python
y
```

```python
x
```

```python
tree=DecisionTreeClassifier(max_depth=3)
tree.fit(x,y)
```

```python
from sklearn.tree import export_graphviz # 트리 출력 함수
dot_data=export_graphviz(tree,out_file=None,feature_names=iris.feature_names[:-2],
               class_names=iris.target_names,rounded=True,filled=True, impurity=True)
```

```python
import pydotplus
graph=pydotplus.graph_from_dot_data(dot_data)
```

```python
graph.write_png('iris.png')

from IPython.core.display import Image
Image('iris.png')
```
