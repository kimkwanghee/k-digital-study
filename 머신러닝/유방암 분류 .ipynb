```python
import pandas as pd
breast_cancer=pd.read_csv('breast-cancer-wisconsin.data')
```

```python
breast_cancer.columns=['id_number','clumb_thickness','unif_cell_size','unif_cell_shape'
                      ,'marg_adesion','single_epith_cell_size','bare_nuciei','bland_chromatin'
                      ,'normal_nuclioli','mitoses','class']
```

```python
breast_cancer['bare_nuciei']=breast_cancer['bare_nuciei'].replace('?',np.nan)
```

```python
breast_cancer['bare_nuciei']=breast_cancer['bare_nuciei'].fillna(breast_cancer['bare_nuciei'].value_counts().index[0])
```

```python
breast_cancer.loc[breast_cancer['class']==4,'cancer_ind']=1
```

```python
x=breast_cancer.drop(['id_number','class','cancer_ind'],axis=1)
y=breast_cancer['cancer_ind']
```

```python
from sklearn.model_selection import train_test_split   #
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler   # 사이킷런에는 표준화를 위한 StandardScaler 변환기
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)  #정규화 한다고 항상 성능이 좋아지는 것은 아님 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)  #k값을 1로 설정 
knn.fit(x_train_scaled,y_train)  # 정규화된 테스트 데이터를 knn알고리즘에 훈련
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, roc_auc_score,roc_curve
y_pred=knn.predict(x_test_scaled)  # 훈련 데이터로 구축된 knn 알고리즘을 정규화된 테스트 데이터로 예측 
```

```python
confusion_matrix(y_test,y_pred)  #오차 행렬-> y가 0과 1로 이루어졌기 때문에 오차행렬 사용 y_pred: 얘측값 y_test: 실제값 
```

```python
accuracy_score(y_test,y_pred)   # 혼동행렬에서 (140+62)/(140+2+6+62)의 값 
```

```python
roc_auc_score(y_test,y_pred)   #roc 커브의 아래쪽 면적을 tpr이 y축 fpr이 x축 
#1에 가까울수록 클래스를 분류하는 성능이 뛰어남 
#즉 무조건 0이라고 예측하여 정확도를 높이는 것에 대한 가능성을 줄여줌 
```

```python
from sklearn.model_selection import GridSearchCV   #모델에 여러개의 파라미터를 만들어 대입시켜주는 함수 
grid_search=GridSearchCV(knn, {'n_neighbors':[1,2,3,4,5]},n_jobs=-1, cv=7, scoring='roc_auc')  
#estimator=knn, 변수값은 딕셔너리 형태로 대입, n_jobs는 사용 컴퓨터 코어 개수 지정 -1값은 모든 코어 사용, cv=교차검증 변수 개수 
grid_search.fit(x_train_scaled,y_train)    #scoring에 따라 정확도 달라짐 
```

```python
grid_search.best_params_
```

```python
knn_best=grid_search.best_estimator_
```

```python
y_pred=grid_search.predict(x_test_scaled)
y_pred
```

```python
y_pred=knn_best.predict(x_test_scaled)

y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
