```python
import pandas as pd
df=pd.read_csv('train.csv')
```

```python
df
```

```python
def check(x):
    if x<10:
        return '유아'
    elif 10<=x<=20:
        return '10대'
    elif 20<=x<=30:
        return '20대'
    elif 30<=x<=40:
        return '30대'
    elif 40<=x<=50:
        return '40대'
    elif 50<=x<=60:
        return '50대'
    elif 50<=x<=60:
        return '60대'
    else:
        return '노인'
```

```python


df['Age_Range']=df['Age'].apply(check)
```

```python
df
```

```python
df.head()
```

```python
result=df.groupby('Age_Range')['Survived'].sum()
```

```python
result
```

```python
result.index
```

```python
result.reindex(['유아','10대', '20대', '30대', '40대', '50대', '노인'])
```

```python
plt.bar(result.index,result,color='g')
plt.title('나이대별 타이타닉 생존자')
plt.xlabel('나이')
plt.ylabel('생존자')
plt.show()
```

```python
result_sex=df.groupby('Sex')['Survived'].sum()
result_sex
```

```python
df=pd.read_csv('kr_daily.csv')
df
```

```python
def datetime(x):
    x=str(x)
    x=x[0:4]+'-'+x[4:6]+'-'+x[6:8]
    return x
```

```python
df.index=df['date'].apply(datetime)
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(df['2020-05-01':].index ,df['2020-05-01':]['confirmed'].values,color='r',linestyle='--')  
plt.xticks(rotation=80)
plt.title('확진자')

plt.subplot(122)
plt.bar(df['2020-05-01':].index, df['2020-05-01':]['death'].values,color='r')
plt.xticks(rotation=80)
plt.title('사망자')
plt.show()
```
