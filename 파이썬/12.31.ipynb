```python
db = pymysql.connect(host='127.0.0.1', port=3306, user='kwanghee', passwd='5000', db='w3schools', charset='utf8')
```

```python
cursor=db.cursor()
```

```python
sql='''
create table scores4(
    id INT(10) NOT NULL AUTO_INCREMENT PRIMARY KEY,
    name varchar(20),
    math int,
    science int,
    english int
)
'''

cursor.execute(sql)
db.commit()
```

```python
while True:
    select_menu=int(input('1-입력, 2-조회, 3-삭제, 0-종료 :'))
    if select_menu==1:
        name=input('이름: ')
        math=int(input('점수: '))
        science=int(input('점수: '))
        english=int(input('점수: '))
        sql='insert into scores4(name,math,science,english) values("{}",{},{},{})'.format(name,math,science,english)
        cursor.execute(sql)
        db.commit()
    elif select_menu==2:
        sql='select * from scores4'
        cursor.execute(sql)
        resultset=cursor.fetchall()
        for result in resultset:
            print('[{}] 이름: {}, 수학: {}, 과학: {}, 영어: {}'.format(result[0],result[1],result[2],result[3],result[4]))
    elif select_menu==3:
        delete_id=int(input('몇번째 id를 삭제하겠습니까?'))
        sql='delete from scores4 where id={}'.format(delete_id)
        cursor.execute(sql)
        db.commit() 
    elif select_menu==0:
        print('종료되었습니다.')
        db.close()
        break
```

```python
cursor=db.cursor()
sql='select * from order_details inner join products on order_details.ProductID=products.ProductID inner join suppliers on products.SupplierID=suppliers.SupplierID order by order_details.Quantity desc limit 3'
cursor.execute(sql)
resultset=cursor.fetchall()
for result in resultset:
    print(result)
```

```python
cursor=db.cursor()
sql='select order_details.Quantity, suppliers.SupplierName from order_details inner join products on order_details.ProductID=products.ProductID inner join suppliers on products.SupplierID=suppliers.SupplierID order by order_details.Quantity desc limit 3'
cursor.execute(sql)
resultset=cursor.fetchall()
for result in resultset:
    print(result)
```

```python
cursor=db.cursor()
sql='''
select order_details.Quantity, categories.categoryid, customers.City from order_details inner join orders on order_details.orderid=orders.orderid
inner join customers on customers.CustomerID=orders.CustomerID inner join products on products.ProductID=order_details.ProductID inner join categories on categories.CategoryID=products.CategoryID
order by  order_details.Quantity desc
'''

cursor.execute(sql)
resultset=cursor.fetchall()
for result in resultset:
    print(result)
```

```python
cursor=db.cursor()
sql='''
select order_details.Quantity, products.productname, customers.country from order_details inner join products on order_details.ProductID=products.ProductID 
inner join orders on orders.OrderID=order_details.OrderID inner join customers on customers.CustomerID=orders.CustomerID where customers.country='USA'
order by order_details.Quantity desc
'''

cursor.execute(sql)
resultset=cursor.fetchall()
for result in resultset:
    print(result)
```
