```python
plt.rc('font',family='Malgun Gothic')
x=np.linspace(0,10,100)
y1=x**2
y2=x**3

plt.plot(x,y1,'r',label='1')
plt.plot(x,y2,'g',label='2')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.04,1))   #     bbox_to_anchor:   
plt.show()
```
