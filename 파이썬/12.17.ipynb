```python
sales_data = {
 '연도':[2015, 2016, 2017, 2018, 2019, 2020],
 '판매량':[103, 70, 130, 160, 190, 230],
 '매출':[500000, 300000, 400000, 550000, 700000, 680000],
 '순이익':[370000, 190000, 300000, 480000, 600000, 590000]
}
```

```python
df=pd.DataFrame(sales_data)
```

```python
df
```

```python
df['순이익률']=df['순이익']/df['매출']*100
```

```python
df
```

```python
def check(n):
    if n>80:
        return '높음'
    else:
        return '낮음'
```

```python
df['순이익률비교']=df['순이익률'].apply(lambda n:'높음'if n>80 else '낮음' )
df #apply에 lambda는 많이 씀  시리즈에만 사용 
```

```python
df['순이익률비교']=df['순이익률'].apply(check)
df
```

```python
for name in df.groupby('순이익률비교'):
    print(type(name))
    print(name)    #낮음과 높음으로 그룹핑
```

```python
df_C_D=pd.DataFrame({'판매월':['1월','2월','3월','4월','5월'],
                    '제품C':[120,230,330,404,100],
                    '제품D':[990,910,820,230,300]})
df_C_D
```

```python
df_left=pd.DataFrame({'key':['A','B','C'],'left':[1,2,3]})
df_left
```

```python
df_right=pd.DataFrame({'key':['A','B','D'],'right':[4,5,6]})
df_right
```

```python
df_left.merge(df_right,how='left',on='key')  
```

```python
df_left.merge(df_right,how='right',on='key')  
```

```python
df_left.merge(df_right,how='inner',on='key')  
```

```python
df_left.merge(df_right,how='outer',on='key')  
```
