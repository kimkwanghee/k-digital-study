```python
import requests
import re
from bs4 import BeautifulSoup

url='https://store.musinsa.com/app/items/lists/00{}'
for i in range(1,11):
    response=requests.get(url.format(i))
    soup=BeautifulSoup(response.text,'html.parser')
    total_list=soup.select('div.boxed-list-wrapper div.li_inner')
    # print(total_list)
    for clothe_info in total_list:
        if clothe_info.select('p.item_title>a'):
            print('브랜드: ',clothe_info.select('p')[0].text)
            print('품명: ',re.sub(r'\[.*?\]','',clothe_info.select('p')[1].text.strip()))
            print('가격: ',clothe_info.select('p')[2].text.strip().split()[-1])
            #         musinsa_dict['품명']=clothe_info.select('div>p.item_title>a')
```

```python
import re
import requests
from bs4 import BeautifulSoup 

url='https://store.musinsa.com/app/items/lists/00{}'
for i in range(1,11):
    response=requests.get(url.format(i))
    soup=BeautifulSoup(response.text,'html.parser')
    result=soup.select('#searchList div.article_info')
#     print(result)
    for j in result:
        if j!=None:
            print('브랜드:',j.select('p')[0].text)
            print('품명: ',re.sub('\[.*?\]','',j.select('p')[1].text.strip()))
            print('가격: ',j.select('p')[2].text.strip().split()[-1])
```
