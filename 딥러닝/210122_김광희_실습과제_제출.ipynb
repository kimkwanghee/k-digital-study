## 실습 1번

```python
import pandas as pd
df=pd.read_csv('input3_out1.csv ')
df
```

```python
import numpy as np
df_input=df.drop(['-4'],axis=1)
df_output=df['-4']

x=np.array(df_input.values.tolist())
y=np.array(df_output.values.tolist())
```

```python
x
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

model = Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))
sgd = optimizers.SGD(lr=0.01)
model.compile(optimizer=sgd ,loss='mse',metrics=['mse'])
model.fit(x,y, batch_size=2, epochs=10, shuffle=False)
```

## 실습 2번

```python
import pandas as pd
df1=pd.read_csv('diabetes.csv')
x=df1.drop(['Outcome'],axis=1)
y=df1['Outcome']
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras import optimizers 

x_array=np.array(x.values.tolist())
y_array=np.array(y.values.tolist())
```

- x값을 8개씩 대입하므로 차원은 8, 시그모이드 함수 사용 
- Sequential의 뜻은 순차적, 연속적이라는 뜻입니다 층을 순차적으로 쌓는 것 
- 경사하강법=gradient descent는  함수 값이 낮아지는 방향으로 독립 변수 값을 변형시켜가면서 최종적으로는 최소 함수 값을 갖도록 하는 독립 변수 값을 찾는 방법이다.
- 옵티마이저=최적화 : 학습속도를 빠르고 안정적이게 하는 것으로 최적의 매개변수를 찾는 것 
- 확률적 경사하강법= sgd -> 옵티마이저의 종류 중 하나로서  (학습률=lr만큼 기울기를 변경시킴 ) -> 데이터를 다 보지 않고 일부만 보고 판단 
- 손실함수 : 실제 값과 예측값의 차이를 측정하는 함수 
- binary_crossentropy : 이진 분류에서 사용하는 손실함수 
- 모델 컴파일 : 모델을 훈련시키기 전에 훈련 과정에 필요한 매개변수를 정의하는 것 


```python
model=Sequential()
model.add(Dense(1, input_dim=8, activation='sigmoid')) 
adagrad=optimizers.Adagrad(lr=0.01)
model.compile(optimizer=adagrad ,loss='binary_crossentropy',metrics=['binary_accuracy'])
model.fit(x_array,y_array, batch_size=5, epochs=300, shuffle=False)
```
